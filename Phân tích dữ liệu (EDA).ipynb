# Phân tích mối quan hệ giữa hai biến feature (Feature-Feature).


1.1. Biến số với biến số (Numerical vs Numerical)

Tương quan giữa Age với Financial Stress

```python
sns.lmplot(data=df, x="Age", y="Financial Stress", line_kws={"color": "red"})
plt.title("Regression Plot: Age với Financial Stress")
plt.xlabel("Age")
plt.ylabel("Financial Stress")
plt.show()
```

Ta thấy:

Tương quan giữa Dietary Habits với Depression

```python
eat_counts = df['Dietary Habits'].value_counts()
threshold = 100
eat_to_keep = eat_counts[eat_counts >= threshold].index

# Sử dụng .assign() để tạo cột tạm thời 'City_grouped' chỉ cho mục đích trực quan hóa
sns.countplot(
    data=df.assign(Eat_grouped = df['Dietary Habits'].apply(lambda x: x if x in eat_to_keep else 'Other')),
    x="Eat_grouped", hue="Depression"
)

plt.title("Count Plot: So sánh Dietary Habits theo nhóm Depresstion")
plt.xlabel("Dietary Habits")
plt.ylabel("Số lượng")
plt.xticks(rotation=90)
plt.show()
```

Ta thấy:
- Nhóm Healthy: Số người không trầm cảm áp đảo so với trầm cảm.
- Nhóm Unhealthy: Số lượng người trầm cảm có phần cao hơn so với Healthy và Moderate.
- Các nhóm khác có kích thước mẫu nhỏ nên được gom vào Other

# Tương quan Đa biến
Vì dữ liệu hầu hết ở dạng phân loại nên cần phải xử lí để thấy được sự tương quan giữa các cột.

Tạm bỏ các cột không liên quan

```python

new_df = df.drop(columns=['id', 'Name', 'City', 'Profession', 'CGPA', 'Degree', 'Dietary Habits','Depression'])
new_df.head()
```

# Encode các cột dạng phân loại
Chuyển Yes/No thành True/False tức 0/1

```python
new_df['Working Professional or Student'] = new_df['Working Professional or Student'].map({'Working Professional': True, 'Student': False})
new_df['Gender'] = new_df['Gender'].map({'Male': True, 'Female': False})
new_df['Family History of Mental Illness'].head()
```

```python
new_df['Family History of Mental Illness'] = new_df['Family History of Mental Illness'].map({'Yes': True, 'No': False})
new_df['Have you ever had suicidal thoughts ?'] = new_df['Have you ever had suicidal thoughts ?'].map({'Yes': True, 'No': False})
new_df['Family History of Mental Illness'].head()
```

# Kết hợp các cột liên quan
Như phân tích lúc đầu, tùy theo người đó `đi học` hay `đi làm` mà các cột `'Academic Pressure', 'Work Pressure', 'Study Satisfaction', 'Job Satisfaction'` có giá trị hay không.
\
Dựa trên điểm chung `Stress` hay `Satisfaction`, ta kết hợp 2 cột với nhau thành `'Aca/Work Pressure','Aca/Work Satisfaction'`

```python
new_df['Aca/Work Pressure'] = new_df['Academic Pressure'].fillna(new_df['Work Pressure'])
new_df['Aca/Work Satisfaction'] = new_df['Study Satisfaction'].fillna(new_df['Job Satisfaction'])
new_df.drop(columns=['Academic Pressure', 'Work Pressure', 'Study Satisfaction', 'Job Satisfaction'], inplace=True)
new_df.head()
```

# Xử lí Sleep Duration

```python
new_df['Sleep Duration'].unique()
```

Chuyển giá trị dạng text sang dạng số cảu cột `Sleep Duration`. Vì 4 giá trị phổ biến áp đảo nên chỉ chuyển 4 giá trị này.
\
Chuyển bằng cách lấy trung bình của khoảng giá trị đó.

```python
mapping = {
    'Less than 5 hours': 4,
    '7-8 hours': 7.5,
    'More than 8 hours': 9,
    '5-6 hours': 5.5,
}

# Apply mapping and set others to NaN
new_df['Sleep Duration'] = new_df['Sleep Duration'].map(mapping)

# Display result
new_df.head()
```
