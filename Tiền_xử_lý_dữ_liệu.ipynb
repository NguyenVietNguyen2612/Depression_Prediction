# Cột Sleep duration

```python
import re

# Chuẩn hóa về nhãn đúng và ánh xạ sang giá trị số
def normalize_and_map_sleep_duration(val):
    if not isinstance(val, str):
        return np.nan
    val = val.lower().strip()

    # Các trường hợp hợp lệ rõ ràng
    if val in ['less than 5 hours', '5-6 hours', '7-8 hours', 'more than 8 hours']:
        pass
    elif 'less' in val and '5' in val:
        val = 'less than 5 hours'
    elif '5' in val and '6' in val:
        val = '5-6 hours'
    elif '6' in val and '7' in val:
        val = '5-6 hours'
    elif '7' in val and '8' in val:
        val = '7-8 hours'
    elif 'more' in val or '8+' in val:
        val = 'more than 8 hours'
    else:
        # Nếu có số giờ lớn hơn 12 → loại bỏ
        hours = list(map(int, re.findall(r'\d+', val)))
        if hours and max(hours) > 12:
            return np.nan
        return np.nan  # không khớp mẫu

    # Ánh xạ sang giá trị số
    mapping = {
        'less than 5 hours': 4.5,
        '5-6 hours': 5.5,
        '7-8 hours': 7.5,
        'more than 8 hours': 8.5
    }
    return mapping.get(val, np.nan)

# Áp dụng
df['Sleep Duration'] = df['Sleep Duration'].apply(normalize_and_map_sleep_duration)

# Xem kết quả
print(df['Sleep Duration'].value_counts(dropna=False))
```

# Phương pháp:
# Chuyển đổi các giá trị có ý nghĩa tương đương sang các mốc 'Less than 5 hours', '7-8 hours', 'More than 8 hours', '5-6 hours' vì các mốc được liệt kê chiếm tần số xuất hiện lớn trong cột và bao quát được dữ liệu
# Những dữ liệu nhiễu không thuộc mốc nào thì chuyển đổi thành NaN để dễ xử lý hơn

# Sau khi chuyển đổi xong thì số lượng giá trị NaN không quá lớn so với tổng số dữ liệu trong cột

# Cột Dietary Habits

```python
def clean_dietary(val):
    if not isinstance(val, str):
        return np.nan
    val = val.strip().lower()

    # Nhóm Moderate
    if (
        "moderate" in val
        or "less than healthy" in val
        or "less healthy" in val
    ):
        return "Moderate"

    # Nhóm Healthy
    if (
        "healthy" in val
        and not any(keyword in val for keyword in ["un", "no", "less"])
    ):
        return "Healthy"
    if val in ["more healthy"]:
        return "Healthy"

    # Nhóm Unhealthy
    if "unhealthy" in val or "no healthy" in val:
        return "Unhealthy"

    # Các giá trị không thuộc nhóm nào
    return np.nan

# Áp dụng làm sạch
df['Dietary Habits'] = df['Dietary Habits'].apply(clean_dietary)

# Kiểm tra kết quả
print(df['Dietary Habits'].value_counts(dropna=False))
```
